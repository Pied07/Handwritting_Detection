```python
%matplotlib inline
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
```

saving sklearn's default load_digits dataset as digits for recognizing handwritting

```python
digits = load_digits()
```

plotting the digits variable for analysing our dataset

```python
digits.data[0]
plt.gray()
for i in range(5):
    plt.matshow(digits.images[i])
```

Marking out target

```python
digits.target[0:5]
```

importing train_test_split for training and testing pourpose

```python
from sklearn.model_selection import train_test_split
```

setting the values

```python
x_train,x_test,y_train,y_test = train_test_split(digits.data,digits.target,test_size=0.2)
```

importing Logistic Regression for decision making and analysing the handwritting

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
```

fitting our logistic model with out training data

```python
model.fit(x_train,y_train)
```

testing the accuracy of our model from the testing data

```python
model.score(x_test,y_test)
```

plotting a random data for manually testing

```python
plt.matshow(digits.images[67])
```

predicting manually of a single data

```python
model.predict([digits.data[67]])
```

predicting manually of handful datas

```python
model.predict(digits.data[0:5])
```

Now, to chgeck how many datas are accurately predicted we will use confusion matrix

```python
y_predicted = model.predict(x_test)
```

printing the confusion matrix

```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test,y_predicted)
cm
```

now for better analysis let's plot the confusion matrix with heatmap

```python
import seaborn as sn
```

plotting heatmap of out confusion matrix

```python
plt.figure(figsize=(10,7))
sn.heatmap(cm,annot=True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
```
